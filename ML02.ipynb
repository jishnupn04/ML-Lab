```python
from sklearn.preprocessing import Binarizer
diabetes_df=pd.read_csv('/content/drive/MyDrive/ML Dataset/diabetes.csv')

bn=Binarizer(threshold=120)

glucose_binarized=bn.transform([diabetes_df['Glucose']])[0]
diabetes_df['Glucose Binarized'] = glucose_binarized
diabetes_df[['Age','Glucose','Glucose Binarized']].head(10)
```

```python
iris_df=pd.read_csv('/content/drive/MyDrive/ML Dataset/Iris.csv')
iris_df['PetalLengthRounded']=np.array(np.round(iris_df['PetalLengthCm']),dtype='int')
iris_df
```

```python
from sklearn.preprocessing import PolynomialFeatures
glucose_pressure=diabetes_df[['Glucose','BloodPressure']]
pf=PolynomialFeatures(degree=2,interaction_only=False,include_bias=False)
res=pf.fit_transform(glucose_pressure)
res

int_features=pd.DataFrame(res,columns=['Glucose','Pressure','Glucose^2','Glucose X Pressure',' Pressure ^2'])
int_features
```

```python
fifa_df=pd.read_csv("/content/drive/MyDrive/ML Dataset/FIFA23_official_data.csv").iloc[:50]
age_bin=np.arange(15,45,5)
bin_names=[1,2,3,4,5]

fifa_df['AgeBin']=pd.cut(fifa_df['Age'],bins=age_bin,right=False)
fifa_df['AgeBinLabel']=pd.cut(fifa_df['Age'],bins=age_bin,labels=bin_names)
fifa_df[['Name','Nationality','Age','AgeBin','AgeBinLabel']]
```
