```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random 
import datetime

no_of_records=30

dates=pd.date_range(start="2023-01-01",end="2023-12-31",periods=no_of_records).date
# print(dates)
price=[round(random.uniform(100,1000),2) for _ in range(no_of_records)]
# print(price)
product_ids=[random.randint(100,999) for _ in range(no_of_records)]
# print(product_ids)
quantity_purchased=[random.randint(1,10) for _ in range(no_of_records)]
serail_nos=[f'SN-{random.randint(100,999) }' for _ in range(no_of_records)]
user_ids=[f'UID-{random.randint(10,20)}' for _ in range(no_of_records)]
user_types=['Retail','Wholesale']
user_class=['Class A','Class B','Class C']
purchase_week=[date.isocalendar()[1] for date in dates]
print(purchase_week)

data={
    'Date':dates,
    'Price':price,
    'Product ID':product_ids,
    'Quantity Purchased':quantity_purchased,
    'Serail NO':serail_nos,
    'User ID':user_ids,
    'User Type':random.choices(user_types,k=no_of_records),
    'User Class':random.choices(user_class,k=no_of_records),
    'Purchase Week':purchase_week
}
print(data)

df=pd.DataFrame(data)
print(df.head())

df.to_csv('Customer_Dataset.csv',index=False)
```

```python
df=pd.read_csv('Customer_Dataset.csv')
print(df)
user_id='UID-14'
user_df=df[df['User ID']==user_id]
plt.plot(user_df['Date'],user_df['Price'])
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'Price Trend of {user_id}')
plt.show()
```

```python
plt.plot(df['Date'],df['Price'])
plt.scatter(user_df['Date'],user_df['Price'],marker='*',c="red",label=f'{user_id}')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title(f'Price Trend of {user_id} over time')
plt.legend()
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot(df['Quantity Purchased'])
plt.title('Distribution of Quantity Purchased')
plt.ylabel('Quantity Purchased')
plt.show()

plt.figure(figsize=(10,6))
plt.boxplot(df['Purchase Week'])
plt.title('Distribution of Purchase Week')
plt.ylabel('Purchase week')
plt.show()
```
