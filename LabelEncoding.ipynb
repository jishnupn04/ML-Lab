### 2. Write a program to Transforming Nominal Features, Transforming Ordinal Features and Encoding Categorical Features using one-hot Encoding Scheme

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
```

```python
fifa_df=pd.read_csv('Datasets/players_fifa23.csv')
fifa_df.head(10)
```

```python
cle=LabelEncoder()
encoded_nationality=cle.fit_transform(fifa_df['Nationality'])
print(encoded_nationality)
country_mappings={}
for i,v in enumerate(cle.classes_):
    country_mappings[i]=v
print(country_mappings)
fifa_df['Countries Encoded']=encoded_nationality
fifa_df[['Name','Nationality','Countries Encoded']].head(10)
```

```python
ale=LabelEncoder()
age_encoded=ale.fit_transform(fifa_df['Age'])
age_mappings={}
for i,v in enumerate(cle.classes_):
    age_mappings[i]=v

fifa_df['Age Encoded']=age_encoded
fifa_df[['Name','Age','Age Encoded']].head()
```

```python
fifa_df['BestPosition'].head()
pos_encoded=pd.get_dummies(fifa_df['BestPosition'],dtype=int)
print(pos_encoded)
pd.concat([fifa_df[['Name','Nationality','Countries Encoded','Age','Age Encoded']],pos_encoded],axis=1)
```
